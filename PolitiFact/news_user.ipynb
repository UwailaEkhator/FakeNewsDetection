```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python

column_names = ['NewsID', 'UserID', 'shares']
df = pd.read_csv('PolitiFactNewsUser.txt', delimiter='\t', names=column_names)
print(df.head())
```

```python

data=[]
with open('PolitiFactNewsUser.txt', 'r') as f:
    for line in f:
        news_id, user_id, shares = map(int, line.strip().split())
        data.append((user_id, news_id, shares))

# Create a directed graph
G = nx.DiGraph()
```

```python
for user_id, news_id, shares in data:
    G.add_edge(user_id, news_id, weight=shares)
```

```python
degree_centrality = nx.degree_centrality(G)
#betweenness_centrality = nx.betweenness_centrality(G)
clustering_coefficient = nx.clustering(G.to_undirected())
diameter = nx.diameter(G) if nx.is_strongly_connected(G) else "Not applicable"

print(f"Degree Centrality: {degree_centrality}")
#print(f"Betweenness Centrality: {betweenness_centrality}")
print(f"Clustering Coefficient: {clustering_coefficient}")
print(f"Diameter: {diameter}")
```

```python
import networkx as nx

# Assuming G is your graph created from the dataset
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"Number of Nodes: {num_nodes}")
print(f"Number of Edges: {num_edges}")

# Calculate edge-to-node ratio
if num_nodes > 0:  # Prevent division by zero
    edge_to_node_ratio = num_edges / num_nodes
    print(f"Edge-to-Node Ratio: {edge_to_node_ratio:.4f}")
else:
    print("No nodes in the graph.")
```
